# Initializing workflow

Importing modules and setting up the data directories

```python
import pandas as pd
import os
import re
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
datadir = '../data/20211209_Bigelow/raw/'
```

## Defining functions

```python
def groupraw(datadir, filetype):
    # finds all the raw files of filetype from a given 
    # data dir and groups them by filestem
    #
    # inputs: 
    #     datadir = full path to directory with the raw files
    #     filetype = extension of raw filetypes e.g. '.sed'  
    # outputs: a dictionary with the filestem as the key
    #          and the list of corresponding individual 
    #          files as the value
    
    allfiles = os.listdir(datadir)
    
    if filetype == '.sed':
        datafiles = [f for f in allfiles if re.search('.sed',f)]
        ind = [re.search('_[0-9]{5}\.sed',f).start() for f in datafiles]
    elif filetype == '.sig':
        datafiles = [f for f in allfiles if re.search('.sig',f)]
        ind = [re.search('_[0-9]{4}\.sig',f).start() for f in datafiles]
    else: 
        print("unknown file type, expect either '.sed' or '.sig'")
        
    fstem = sorted(set([f[:ii] for f,ii in zip(datafiles,ind)]))
    filegroups = {}
    for fs in fstem:
        filegroups[fs] = [datadir+f for f in datafiles if re.match(fs+'_[0-9]',f)]
    
    return filegroups
```

# Processing files

Now we can loop through the files, import the data, and combine it into one data frame.

We have to do this different compared to the spectral library, because we don't want to take the mean of the different salad toss scans, we want to keep them all.

Note, I'm also including a regular asco and fucus sample for comparison what the variance is in those cases

```python
filegroups = groupraw(datadir,'.sig')

groupkeys = ['asco_salad','asco4','fucus_salad','fucus4']

data = []
for gkey in groupkeys:
    filelist = filegroups[gkey]
    for ff in filelist:
        fid = re.search('_\d{4}\.sig',ff).group()[1:5]
        data += [pd.read_csv(ff,sep='\s+',skiprows=27,header=None,usecols=[3],names = [gkey+'_'+fid])]

lcdf = pd.concat(data,axis=1)
lcdf['Wvl'] = pd.read_csv(ff,sep='\s+',skiprows=27,header=None,usecols=[0],names = ['Wvl'])
lcdf.set_index('Wvl',inplace = True)
```
